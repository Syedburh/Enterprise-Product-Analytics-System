```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales=pd.read_csv("C:/Users/Inaya/Downloads/Products.csv")
```

```python
sales.head(10)
```

```python
products=pd.read_csv("C:/Users/Inaya/Downloads/sales.csv")
```

```python
products.head(10)
```

```python
1️⃣ Check missing values
```

```python
sales.isnull().sum()
```

```python
products.isnull().sum()
```

```python
Handle missing values
```

```python
sales['DiscountApplies']=sales['DiscountApplied'].fillna(0)
```

```python
sales['QuantitySold']=sales['QuantitySold'].fillna(1)
```

```python
sales = sales.drop('DiscountApplied.1', axis=1)
```

```python
sales=sales.drop('DiscountApplies',axis=1)
```

```python
2️⃣ Convert % to number (critical – you faced this already)
```

```python
sales['DiscountApplied']=sales['DiscountApplied'].str.replace('%',' ').astype(float)
```

```python
sales.head(2)
```

```python
3️⃣ Convert date column
```

```python
sales['Date']=pd.to_datetime(sales['Date'])
```

```python
sales = sales.merge(
    products[['ProductID', 'UnitPrice_INR']],
    on='ProductID',
    how='left'
)
```

```python
4️⃣ Create Revenue column
```

```python
sales['revenue']=sales['QuantitySold']*sales['UnitPrice_INR']*(1-sales['DiscountApplied']/100)
```

```python
sales['revenue']
```

```python
1️⃣ Monthly Sales Trend
```

```python
Monthly_sales=sales.groupby(sales['Date'].dt.to_period('M'))['revenue'].sum()
```

```python
Monthly_sales.plot(figsize=(10,5), title="Monthly Revenue Trend")
```

```python
sales=sales.merge(products[['ProductID','ProductName']],
on='ProductID',
how='left')
```

```python
2️⃣ Top 10 Products by Revenue
```

```python
Top_10_products=sales.groupby('ProductName')['revenue'].sum().sort_values(ascending=False).head(10)
```

```python
Top_10_products
```

```python
Top_10_products.plot(kind='bar',figsize=(10,5),title="Top_10_products",color='red')
```

```python
3️⃣ Discount vs Revenue
```

```python
discount=sales.groupby('DiscountApplied')['revenue'].sum()
```

```python
discount
```

```python
discount.plot(kind='bar',figsize=(10,5),title="Discount impact")
```

```python
STEP 4: Forecasting (Simple & ACCEPTED)
```

```python
monthly_df = Monthly_sales.reset_index()
```

```python
monthly_df = Monthly_sales.reset_index()
monthly_df['Date'] = monthly_df['Date'].astype(str)
monthly_df['Date'] = pd.to_datetime(monthly_df['Date'])
```

```python
Simple forecasting using rolling average
```

```python
monthly_df['Forecast'] = monthly_df['revenue'].rolling(window=3).mean()

monthly_df.plot(x='Date', y=['revenue','Forecast'], figsize=(10,5))
```

```python
STEP 5: Customer Segmentation (RFM – BIG PLUS)
```

```python
rfm = sales.groupby('CustomerID').agg({
    'Date': lambda x: (sales['Date'].max() - x.max()).days,
    'SaleID': 'count',
    'revenue': 'sum'
})

```
